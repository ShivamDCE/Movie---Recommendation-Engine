```python
data=pd.merge(ratings,movies,on='movieId')
```

```python
data.head()
```

```python
data=data.drop(['timestamp','genres'],axis=1)
```

```python
data
```

```python
#Creating a pivot table
data_pivot=pd.pivot_table(data,index='userId',columns='title',values='rating')
```

```python
userRatings = data_pivot.dropna(thresh=10, axis=1).fillna(0,axis=1)
```

```python
userRatings
```

```python
def standardize(row):
    return (row-row.mean()/row.max()-row.min())
    
userRatings=userRatings.apply(standardize).T
```

```python
userRatings
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
corrMatrix=pd.DataFrame(cosine_similarity(userRatings),index=userRatings.index,columns=userRatings.index)
```

```python
corrMatrix
```

```python
corrMatrix['10,000 BC (2008)'].sort_values(ascending=False)
```

```python
def get_similar(movie,rating):
    if rating>2.5:
        return corrMatrix[movie].sort_values(ascending=False).head(50)
    else:
        return corrMatrix[movie]*(-1).sort_values(ascending=False).head(50)
```

```python
get_similar('2 Fast 2 Furious (Fast and the Furious 2, The) (2003)',5.0)
```
